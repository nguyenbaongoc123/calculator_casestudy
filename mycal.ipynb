```python
import math
listm = ['MC','MR','MS','M+','M-']
listc = ['<-','CE','C','+/-','sqrt']
list3 = ['7','8','9','/','%']
list4 = ['4','5','6','*','1/x']
list5 = ['1','2','3','-','=']
list6 = ['0','.','+']
trans= {'--':'+',
       '+-':'-'}
sym = ['*','/','-','+']
import tkinter as tk
from tkinter import ttk

window = tk.Tk()
window.title("Calculator")
window.resizable(width=False, height=False)

listbtn=[]
memory='0'
c=0
ms=0
result=0
##ham tinh total
def cal_res(memory):
    for key in trans.keys():
        if key in memory:
            memory = memory.replace(key,trans[key])  
    new=''
    for char in memory:
        if char.isdigit() or char=='.':
            new+=char
        else:
            new+=' '+char+' '
    nlchar=list(new.split())
    memory=nlchar.copy()    
    
    i=0
    while i<len(memory):
        
        if memory[i]=='/'and memory[i+1]!='-':
            memory[i]=float(memory[i-1])/float(memory[i+1])
            memory.pop(i+1)
            memory.pop(i-1)
        elif memory[i]=='/'and memory[i+1]=='-':
            memory[i]=0-float(memory[i-1])/float(memory[i+2])
            memory.pop(i+2)
            memory.pop(i+1)
            memory.pop(i-1)
        elif memory[i]=='*'and memory[i+1]!='-'   :
            memory[i]=float(memory[i-1])*float(memory[i+1])
            memory.pop(i+1)
            memory.pop(i-1)
        elif memory[i]=='*'and memory[i+1]=='-':
            memory[i]=0-float(memory[i-1])*float(memory[i+2])
            memory.pop(i+2)
            memory.pop(i+1)
            memory.pop(i-1)
        else:
            i+=1
    if memory[0]=='-' and memory[1]=='-':
        re=float(memory[2])
    elif memory[0]!='-':  
        re=float(memory[0])
       
    else:
        re=0-float(memory[1])
    for i in range(1,len(memory)):
        if memory[i]=='+':
            re+=float(memory[i+1])
        elif memory[i]=='-'   :
            re=re-float(memory[i+1])
    return re

def before_num(memory,sign):
    for i in range(len(memory)):
        if memory[i] == sign:
            index=i-1
    new=''
    for char in memory:
        if char.isdigit() or char=='.':
            new+=char
        else:
            new+=' '+char+' '
    nlchar=list(new.split())
    print(nlchar)
    #memory=nlchar.copy()      
    for i in range(len(nlchar)):
        if nlchar[i] == sign:
            num=nlchar[i-1]
    
    return num,index


############33333333333333333333333333333333333
def callback(e):
    global memory
    global ms
    global result
    label['text']=memory
    ## '='
    if e.widget['text']=="=":
        result= cal_res(memory)
        print("total:", result)
        label['text']+='='
        labelre['text']=result
        memory=str(result)
        #print(memory)
    elif e.widget['text']=="MC":
        ms=0
    elif e.widget['text']=="MS":
        ms=cal_res(memory)
    elif e.widget['text']=="MR":
        mr=ms
        label['text']=mr
        labelre['text']=mr
        memory=str(mr)
    elif e.widget['text']=="M+":
        pass
        ms+=cal_res(memory)
    elif e.widget['text']=="M-":
        pass
        ms-=cal_res(memory)
    elif e.widget['text']=="<-":
        label['text']=label['text'][:-1] 
        memory=label['text']
    # CE
    
    elif e.widget['text']=="CE":
        memory='0'
        label['text']=0
        labelre['text']=0
    elif e.widget['text']=="C":
        memory='0'
        label['text']=0
        labelre['text']=0    
    #SQRT
    elif e.widget['text']=="sqrt":
        sign="s"
        memory+='s'
        
        num,index=before_num(memory,sign)
        print("index ",index)
        print("mem truoc thai doi " ,memory)
        label['text']=f'sqrt({num})'
        labelre['text']=math.sqrt(float(num))
        memory = memory.replace(str(num+sign),str(labelre['text']))
        
        #label['text']=memory
        print("mem sau thai doi " ,memory)
    # 1/X
    elif e.widget['text']=="1/x":
        if memory == '0':
            labelre['text']="Cannot divide by zero"
        else:    
            sign="s"
            memory+='s'
            num,index=before_num(memory,sign)
            print("mem truoc thai doi " ,memory)
            label['text']=f'1/{num}'
            labelre['text']=f'1/{num}'
            memory =  memory.replace(str(num+sign),str(labelre['text']))
        
        print("mem sau thai doi " ,memory)
 
    # +/-
    elif e.widget['text']=="+/-":
            
            sign="n"
            memory+='n'        
            num ,index=before_num(memory,sign)
            print("index ",index)
            print("mem truoc thai doi " ,memory)
            labelre['text']=f'-{num}'
            memory =  memory.replace(str(num+sign),str(labelre['text']))
            label['text']=memory
            print("mem sau thai doi " ,memory)

    # %
    elif e.widget['text']=="%":
        sign="f"
        memory+='f'        
        num,index=before_num(memory,sign)
        print("num: ", num)
        print("index ",index)
        print("mem truoc thai doi " ,memory)
        label['text']=f'{num}%'
        labelre['text']=f'{num}/100'
        memory =  memory.replace(str(num+sign),str(labelre['text']))
       
        print("mem sau thai doi " ,memory)
    ## dang int

    elif e.widget['text'].isdigit():
        if memory=='0' or memory == str(result):
            label['text']=e.widget['text']
        elif  memory!='0' and memory != str(result):
            label['text']+=e.widget['text']
            
        memory=label['text']
    else:
        if e.widget['text'] in sym:
            if label['text'][-1] in sym:
                label['text']=label['text'][:-1]+e.widget['text']
            else:
                label['text']+=e.widget['text']
                memory=label['text']

        
        
        
        
frame=tk.Frame(window,borderwidth=2, relief="groove", bg='white', width=140, height=10) 
#frame.resizable(width=False, height=False)
for i in range(7):
        if i ==0:
            for j in range(5):
                frame.grid(row=0, columnspan=7,padx=5, pady=5, ipadx=1, ipady=15)
                label=tk.Label(frame, text='0',anchor='e',justify='right',bg='white', font = ('Helvetice',16),width=25)
                label.grid(row=0,columnspan=7 )
                labelre=tk.Label(frame, text='0',anchor='e',justify='right',bg='white', font = ('Helvetice',18,'bold'),width=18)
                labelre.grid(row=1,columnspan=7)
        elif i==1:
            for j in range(5):
                btn =tk.Button(window, text =listm[j], width=5)
                listbtn.append(btn)
                listbtn[c].grid(row=i, column=j,ipadx=3, ipady=8, padx=4, pady=4)
                btn.bind("<Button-1>",callback)
                c+=1
        elif i==2:
            for j in range(5):
                btn =tk.Button(window, text =listc[j], width=5)
                listbtn.append(btn)
                listbtn[c].grid(row=i, column=j,ipadx=3, ipady=8, padx=4, pady=4)
                btn.bind("<Button-1>",callback)
              
                c+=1
        elif i==3:
            for j in range(5):
                btn =tk.Button(window, text =list3[j], width=5)
                listbtn.append(btn)
                listbtn[c].grid(row=i, column=j,ipadx=3, ipady=8, padx=4, pady=4)  
                btn.bind("<Button-1>",callback)
                
                c+=1
        elif i==4:
            for j in range(5):
                btn =tk.Button(window, text =list4[j], width=5)
                listbtn.append(btn)
                listbtn[c].grid(row=i, column=j,ipadx=3, ipady=8, padx=4, pady=4)
                btn.bind("<Button-1>",callback)
                c+=1
        elif i==5:
            for j in range(5):
                if j==4:
                    btn =tk.Button(window, text =list5[j], width=5)
                    listbtn.append(btn)
                    listbtn[c].grid( row=i, column=j,ipadx=3, ipady=32, padx=4, pady=4, rowspan=2)
                    btn.bind("<Button-1>",callback)
                    c+=1
                else:
                    btn =tk.Button(window, text =list5[j], width=5)
                    listbtn.append(btn)
                    listbtn[c].grid(row=i, column=j,ipadx=3, ipady=8, padx=4, pady=4)
                    btn.bind("<Button-1>",callback)
                    c+=1
        elif i==6:
            for j in range(3):
                if j==0:
                    btn =tk.Button(window, text =list6[j], width=5)
                    listbtn.append(btn)
                    listbtn[c].grid(row=i, column=j,ipadx=32, ipady=8, padx=4, pady=4 , columnspan=2)
                    btn.bind("<Button-1>",callback)
                    c+=1
                else:
                    btn =tk.Button(window, text =list6[j], width=5)
                    listbtn.append(btn)
                    listbtn[c].grid(row=i, column=j+1,ipadx=3, ipady=8, padx=4, pady=4)
                    btn.bind("<Button-1>",callback)
                    c+=1
## handle event

window.mainloop()

```
